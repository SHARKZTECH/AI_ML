```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

model_name = "deepset/tinyroberta-squad2"

# Load model and tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Set up the question-answering pipeline
nlp = pipeline('question-answering', model=model, tokenizer=tokenizer)

quit_command = "quit"
while True:
    # Get user input
    question = input("Enter your question (type 'quit' to exit): ")
    if question.lower() == quit_command:
        print("Exiting...")
        break

    context = input("Enter the context: ")

    # Perform question answering
    QA_input = {
        'question': question,
        'context': context
    }
    res = nlp(QA_input)

    # Print the answer
    answer = res['answer']
    print(f"Answer: {answer}")
    print()

```

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

model_name = "deepset/tinyroberta-squad2"

# Load model and tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Set up the question-answering pipeline
nlp = pipeline('question-answering', model=model, tokenizer=tokenizer)

quit_command = "quit"

# Load the context from the file
filename = "context.txt"
with open(filename, "r") as file:
    context_text = file.read().replace('.', ',')

while True:
    # Get user input
    question = input("Enter your question (type 'quit' to exit): ")
    if question.lower() == quit_command:
        print("Exiting...")
        break

    # Perform question answering with the loaded context
    res = nlp(question=question, context=context_text)

    # Check if the model provided an answer with confidence
    if res['score'] > 0.1:
        answer = res['answer']
        print(f"Answer: {answer}")
    else:
        print("Answer: Not available")

    print()  # Add a line break between iterations
```
