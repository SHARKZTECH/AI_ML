```python
df = pd.read_csv('country_wise_latest.csv')
```

# 1. Scatter plot
 We can use a scatter plot to visualize the relationship between two variables, such as the number of confirmed cases and deaths:

```python
plt.scatter(df['Confirmed'], df['Deaths'])
plt.xlabel('Confirmed cases')
plt.ylabel('Deaths')
plt.title('COVID-19: Confirmed cases vs. Deaths')
plt.show()
```

# 2. Bar plot
We can use a bar plot to compare the number of cases or deaths across different countries or regions:

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='Confirmed', data=df)
plt.xticks(rotation=90)
plt.xlabel('Country/Region')
plt.ylabel('Confirmed cases')
plt.title('COVID-19: Confirmed cases by country/region')
plt.show()
```

# 3. Heatmap
We can use a heatmap to visualize the correlation between different variables:

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('COVID-19: Correlation heatmap')
plt.show()
```

```python
# Create a plot using Seaborn
sns.barplot(x='Country/Region', y='Confirmed', data=df)

# Customize the plot
plt.xticks(rotation=90)
plt.xlabel('Country/Region')
plt.ylabel('Confirmed cases')
plt.title('COVID-19: Confirmed cases by country/region')

# Set the DPI (dots per inch) to 300
plt.figure(dpi=300)

# Save the plot as a PNG file
plt.savefig('covid19_infographic.png', bbox_inches='tight')
```

```python
# Create a figure with a 2:1 aspect ratio
fig = plt.figure(figsize=(12, 6))

# Set up a grid with two rows and three columns using gridspec()
gs = fig.add_gridspec(nrows=2, ncols=3, width_ratios=[3, 3, 4])

# Create the first plot in the top-left grid
# Create the first plot in the top-left grid
ax1 = fig.add_subplot(gs[0, 0])
# Sort the dataframe by the number of confirmed cases
df_sorted = df.sort_values(by='Confirmed', ascending=False)
# Select the top 10 countries
df_top10 = df_sorted.head(10)
sns.barplot(x='Country/Region', y='Confirmed', data=df_top10, ax=ax1)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
ax1.set_title('Top 10 countries by confirmed cases')

# Create the second plot in the top-center grid
# Create the plot in the top-right grid
ax2 = fig.add_subplot(gs[0, 1])
# Sort the dataframe by the number of deaths
df_sorted = df.sort_values(by='Deaths', ascending=False)
# Select the top 10 countries
df_top10 = df_sorted.head(10)
sns.barplot(x='Country/Region', y='Deaths', data=df_top10, ax=ax2)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')
ax2.set_title('Top 10 countries by deaths')


# Create the third plot in the top-right grid
ax3 = fig.add_subplot(gs[0, 2])
sns.lineplot(x='1 week change', y='Confirmed', data=df, ax=ax3)
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=45, ha='right')
ax3.set_title('Confirmed cases over time')

# Create the fourth plot in the bottom-left grid
ax4 = fig.add_subplot(gs[1, 0])
sns.lineplot(x='1 week change', y='Deaths', data=df, ax=ax4)
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=45, ha='right')
ax4.set_title('Deaths over time')

# Create the fifth plot in the bottom-center grid
ax5 = fig.add_subplot(gs[1, 1])
# Sort the dataframe by the percent increase in deaths over the past week
df_sorted = df.sort_values(by='1 week % increase', ascending=False)
# Select the top 10 countries
df_top10 = df_sorted.head(10)
sns.barplot(x='1 week % increase', y='Confirmed', data=df_top10, ax=ax5)
ax5.set_xticklabels(ax5.get_xticklabels(), rotation=45, ha='right')
ax5.set_title('Confirmed cases by severity')

# Create the sixth plot in the bottom-right grid
# Create the plot in the bottom-right grid
ax6 = fig.add_subplot(gs[1, 2])
# Sort the dataframe by the percent increase in deaths over the past week
df_sorted = df.sort_values(by='1 week % increase', ascending=False)
# Select the top 10 countries
df_top10 = df_sorted.head(10)
sns.barplot(x='1 week % increase', y='Deaths', data=df_top10, ax=ax6)
ax6.set_xticklabels(ax6.get_xticklabels(), rotation=45, ha='right')
ax6.set_title('Top 10 countries by deaths and severity')


# Adjust the spacing between the plots
plt.subplots_adjust(hspace=1, wspace=0.5)

# Save the figure as a PNG file with a resolution of 300 dpi
plt.savefig('your_student_id_dashboard.png', dpi=300, bbox_inches='tight')
```
